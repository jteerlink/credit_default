```python
val_df = pd.read_csv('../data/val_predictions.csv')
```

```python
def calculate_lift_by_deciles(df):
    """
    Calculate lift by deciles for a binary classification model
    
    Parameters:
    df: pandas Dataframe with y_true and y_pred columns

    Returns:
    DataFrame with decile analysis
    """
    
    # Create deciles based on predicted probability (10 = highest risk, 1 = lowest risk)
    df['decile'] = pd.qcut(df['y_pred'], q=10, labels=False, duplicates='drop') + 1
    
    # Calculate statistics by decile
    decile_stats = df.groupby('decile').agg({
        'y_true': ['count', 'sum', 'mean'],
        'y_pred': ['mean', 'min', 'max']
    }).round(4)
    
    # Flatten column names
    decile_stats.columns = ['total_count', 'defaulters', 'default_rate', 
                           'avg_pred_prob', 'min_pred_prob', 'max_pred_prob']
    
    # Calculate overall default rate (baseline)
    overall_default_rate = df['y_true'].mean()
    
    # Calculate lift (how much better than random)
    decile_stats['lift'] = decile_stats['default_rate'] / overall_default_rate
    
    # Calculate cumulative statistics
    decile_stats = decile_stats.sort_index()
    decile_stats['cum_defaulters'] = decile_stats['defaulters'].cumsum()
    decile_stats['cum_population'] = decile_stats['total_count'].cumsum()
    decile_stats['cum_default_rate'] = decile_stats['cum_defaulters'] / decile_stats['cum_population']
    
    # Calculate gains (cumulative % of defaulters captured)
    total_defaulters = df['y_true'].sum()
    decile_stats['gains'] = (decile_stats['cum_defaulters'] / total_defaulters) * 100
    
    return decile_stats, overall_default_rate
```

```python
decile_stats, overall_default_rate = calculate_lift_by_deciles(val_df)
```

```python
sns.barplot(decile_stats, x=decile_stats.index,y=decile_stats['lift'])
plt.axhline(y=1, color='red', linestyle='--', linewidth=2, alpha=0.7)
plt.title('Lift by Decile vs Random Selection', fontweight='bold')
plt.show()
```

```python
sns.barplot(decile_stats, x=decile_stats.index, y=decile_stats['default_rate'] * 100, color='green', alpha=0.8, 
            edgecolor='black', linewidth=0.5)
plt.axhline(y=overall_default_rate * 100, color='red', linestyle='--', linewidth=2, 
            alpha=0.7, label=f'Overall Rate: {overall_default_rate:.1%}')
plt.xlabel('Risk Decile', fontweight='bold')
plt.ylabel('Default Rate (%)', fontweight='bold')
plt.title('Actual Default Rate by Decile', fontweight='bold')
plt.legend()
plt.show()
```

```python
non_defaulters = val_df[val_df['y_true']==0]['y_pred']
defaulters = val_df[val_df['y_true']==1]['y_pred']


plt.hist(non_defaulters, bins=50, alpha=0.7, label='Non-Defaulters', 
         color='green', density=True)
plt.hist(defaulters, bins=50, alpha=0.7, label='Defaulters', 
         color='red', density=True)

plt.xlabel('Predicted Probability', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Distribution of Predicted Probabilities by Actual Class', 
          fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

# Add vertical lines for means
plt.axvline(non_defaulters.mean(), color='green', linestyle='--', alpha=0.8,
            label=f'Non-Default Mean: {non_defaulters.mean():.3f}')
plt.axvline(defaulters.mean(), color='red', linestyle='--', alpha=0.8,
            label=f'Default Mean: {defaulters.mean():.3f}')

plt.legend()
plt.tight_layout()                                                                                                                                              
```
