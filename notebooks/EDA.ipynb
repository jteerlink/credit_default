# Distributions, Correlations, and Outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt
for col in raw_df.columns:
    sns.histplot(raw_df[col], kde=True)
    plt.show()
```

```python
continuous_vars = raw_df.select_dtypes(include=['number']).columns.tolist()
```

```python
aggregated_decile_df = pd.DataFrame()


for var in continuous_vars:
    # Drop NaN values for the current continuous variable and the defaulted_col for qcut
    subset = raw_df.dropna(subset=[var, 'DefaultedAdvances']).copy()

    if not subset.empty and len(subset[var].unique()) >= 10: # qcut needs at least q unique values
        # Calculate decile bins for the current continuous variable
        # qcut returns the bin labels (0-9 by default, so we add 1 for 1-10)
        subset['decile_bin'] = pd.qcut(subset[var], q=10, labels=False, duplicates='drop') + 1

        # Calculate the mean of defaulted_col for each decile bin of 'var'
        # This gives the proportion of defaulted advances in each decile of 'var'
        decile_agg = subset.groupby('decile_bin')['DefaultedAdvances'].mean().reset_index()
        decile_agg.rename(columns={'DefaultedAdvances': 'default_rate'}, inplace=True)

        # Add column for the continuous variable name
        decile_agg['continuous_variable'] = var

        aggregated_decile_df = pd.concat([aggregated_decile_df, decile_agg])
    elif len(subset[var].unique()) < 10:
        print(f"Warning: Not enough unique values in '{var}'. Skipping.")

```

```python
num_plots = len(continuous_vars)
fig, axes = plt.subplots(num_plots, 1, figsize=(12, 7 * num_plots))

for i, var in enumerate(continuous_vars):
    ax = axes[i]
    plot_data = aggregated_decile_df[aggregated_decile_df['continuous_variable'] == var]

    if not plot_data.empty:
        sns.lineplot(
            data=plot_data,
            x='decile_bin',
            y='default_rate',
            ax=ax,
            linewidth=2.5,
            marker='o',
            color='skyblue'
        )
        ax.set_title(f'Default Rate by Decile of {var}', fontsize=14)
        ax.set_xlabel(f'Decile of {var}', fontsize=12)
        ax.set_ylabel('Proportion of Defaulted Advances', fontsize=12)
        ax.set_xticks(range(1, 11))
```

### Outlier Detection

Using a zscore approach where |zscore| > 3 = outlier.  I like this approach since it identifies more extreme cases and still want to capture majority of variance where possible especially for GBM models.  If using a logistic regression I would likely floor and cap values.

```python
from scipy.stats import zscore

# Calculate Z-score for each numerical column
for col in continuous_vars:
    raw_df[f'Z_score_{col}'] = zscore(raw_df[col], nan_policy='omit')

threshold = 3

# Identify outliers
outliers = {}
for col in continuous_vars:
    outlier_rows = raw_df[(raw_df[f'Z_score_{col}'] > threshold) | (raw_df[f'Z_score_{col}'] < -threshold)]
    num_outliers = len(outlier_rows)
    total_rows = len(raw_df)
    percentage_outliers = (num_outliers / total_rows) * 100 if total_rows > 0 else 0
    outliers[col] = {
        'count': num_outliers,
        'percentage': percentage_outliers
    }

outliers_df = pd.DataFrame.from_dict(outliers, orient='index')
outliers_df
```
